```python
# Brain Metastases Early Detection using CT-MRI Fusion
# Extended from: A Deep Hybrid Framework for Multimodal Medical Image Fusion
# Target: Early detection of brain metastases using enhanced fusion techniques

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as patches
from matplotlib import cm
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from PIL import Image
import nibabel as nib
import pandas as pd
from sklearn.model_selection import train_test_split
import torchvision.models as models
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import scipy.ndimage as ndi
from sklearn.ensemble import RandomForestClassifier
import joblib
import logging
import glob
from scipy.ndimage import zoom
import json

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# ==================== ENHANCED FUSION MODEL ====================

class AttentionBlock(nn.Module):
    """Attention mechanism for better feature selection"""
    def __init__(self, in_channels):
        super(AttentionBlock, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, in_channels//4, 1)
        self.conv2 = nn.Conv2d(in_channels//4, in_channels, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        attention = self.conv1(x)
        attention = torch.relu(attention)
        attention = self.conv2(attention)
        attention = self.sigmoid(attention)
        return x * attention

class EnhancedFusionAutoencoder(nn.Module):
    """Enhanced autoencoder with attention mechanism for brain metastases detection"""
    def __init__(self):
        super(EnhancedFusionAutoencoder, self).__init__()

        # Encoder for CT images
        self.ct_encoder = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            AttentionBlock(128),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            AttentionBlock(256),
        )

        # Encoder for MRI images
        self.mri_encoder = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            AttentionBlock(128),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            AttentionBlock(256),
        )

        # Fusion layer with adaptive weights
        self.fusion_conv = nn.Sequential(
            nn.Conv2d(512, 256, 3, padding=1),
            nn.ReLU(),
            AttentionBlock(256),
            nn.Conv2d(256, 128, 3, padding=1),
            nn.ReLU(),
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(128, 128, 2, stride=2),
            nn.ReLU(),
            AttentionBlock(128),
            nn.ConvTranspose2d(128, 64, 2, stride=2),
            nn.ReLU(),
            nn.Conv2d(64, 32, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 1, 3, padding=1),
            nn.Sigmoid()
        )

    def forward(self, ct_img, mri_img):
        # Extract features from both modalities
        ct_features = self.ct_encoder(ct_img)
        mri_features = self.mri_encoder(mri_img)

        # Concatenate features
        fused_features = torch.cat([ct_features, mri_features], dim=1)

        # Apply fusion convolution
        fused_features = self.fusion_conv(fused_features)

        # Decode to final image
        fused_image = self.decoder(fused_features)

        return fused_image

class MetastasesClassifier(nn.Module):
    """Classification network for metastases detection"""
    def __init__(self, num_classes=2):  # 0: No metastases, 1: Metastases present
        super(MetastasesClassifier, self).__init__()

        # Load pre-trained VGG19 for feature extraction
        self.vgg19 = models.vgg19(pretrained=True).features

        # Freeze VGG19 parameters initially
        for param in self.vgg19.parameters():
            param.requires_grad = False

        # Custom classifier for metastases detection
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((7, 7)),
            nn.Flatten(),
            nn.Linear(512 * 7 * 7, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # Extract features using VGG19
        features = self.vgg19(x)

        # Classify
        output = self.classifier(features)
        return output

# ==================== DATASET HANDLERS ====================

class BrainMetastasesDataset(Dataset):
    """Dataset class for brain metastases detection"""
    def __init__(self, ct_dir, mri_dir, labels_file=None, transform=None, mode='fusion'):
        self.ct_dir = ct_dir
        self.mri_dir = mri_dir
        self.transform = transform
        self.mode = mode

        # Get all CT files
        self.ct_files = sorted([f for f in os.listdir(ct_dir) if f.endswith(('.png', '.jpg', '.nii', '.nii.gz'))])
        self.mri_files = sorted([f for f in os.listdir(mri_dir) if f.endswith(('.png', '.jpg', '.nii', '.nii.gz'))])

        # Load labels if provided
        self.labels = None
        if labels_file and os.path.exists(labels_file):
            self.labels = pd.read_csv(labels_file)

    def __len__(self):
        return min(len(self.ct_files), len(self.mri_files))

    def load_image(self, file_path):
        """Load image (supports both regular images and NIfTI files)"""
        if file_path.endswith('.nii') or file_path.endswith('.nii.gz'):
            # Load NIfTI file
            nii_img = nib.load(file_path)
            img_data = nii_img.get_fdata()

            # If 3D, take middle slice
            if len(img_data.shape) == 3:
                img_data = img_data[:, :, img_data.shape[2]//2]

            # Normalize to 0-255 range
            img_data = ((img_data - img_data.min()) / (img_data.max() - img_data.min()) * 255).astype(np.uint8)
            return img_data
        else:
            # Load regular image
            img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
            return img

    def __getitem__(self, idx):
        # Load CT and MRI images
        ct_path = os.path.join(self.ct_dir, self.ct_files[idx])
        mri_path = os.path.join(self.mri_dir, self.mri_files[idx])

        ct_img = self.load_image(ct_path)
        mri_img = self.load_image(mri_path)

        # Resize to standard size
        ct_img = cv2.resize(ct_img, (128, 128))
        mri_img = cv2.resize(mri_img, (128, 128))

        # Convert to tensor
        ct_tensor = torch.FloatTensor(ct_img).unsqueeze(0) / 255.0
        mri_tensor = torch.FloatTensor(mri_img).unsqueeze(0) / 255.0

        if self.mode == 'fusion':
            return ct_tensor, mri_tensor
        elif self.mode == 'classification':
            # For classification, we need labels
            label = 0  # Default: no metastases
            if self.labels is not None:
                # Get label from CSV file by matching filename or index
                filename_base = self.ct_files[idx].replace('.png', '').replace('ct_', '')
                matching_row = self.labels[self.labels['filename'].str.contains(filename_base, na=False)]
                if not matching_row.empty:
                    label = int(matching_row.iloc[0]['has_metastases'])
                else:
                    # If no match found, use index-based lookup
                    if idx < len(self.labels):
                        label = int(self.labels.iloc[idx]['has_metastases'])

            return ct_tensor, mri_tensor, torch.LongTensor([label])

# ==================== REAL DATASET INTEGRATION ====================

def download_medical_decathlon_brain_data(data_dir='./medical_decathlon_data'):
    """Download and setup Medical Segmentation Decathlon Brain Dataset"""
    import requests
    import tarfile
    from tqdm import tqdm

    os.makedirs(data_dir, exist_ok=True)

    print("=== MEDICAL SEGMENTATION DECATHLON - BRAIN DATASET ===")
    print("This function will help you download the real brain tumor dataset.")
    print("Dataset contains: T1, T1Gd, T2, FLAIR MRI sequences + segmentation masks")
    print("Size: ~2GB")

    # Multiple download options
    download_urls = {
        'aws_direct': 'https://msd-for-monai.s3-us-west-2.amazonaws.com/Task01_BrainTumour.tar',
        'academic_torrents': 'https://academictorrents.com/download/274be65156ed14828fb7b30b82407a2417e1924a.torrent'
    }

    print("\nDOWNLOAD OPTIONS:")
    print("1. AWS Direct Download (Recommended)")
    print("2. Academic Torrents")
    print("3. Manual Download Instructions")

    choice = input("Select option (1/2/3): ").strip()

    if choice == '1':
        print("Downloading from AWS...")
        url = download_urls['aws_direct']
        filename = os.path.join(data_dir, 'Task01_BrainTumour.tar')

        try:
            response = requests.get(url, stream=True)
            total_size = int(response.headers.get('content-length', 0))

            with open(filename, 'wb') as f, tqdm(
                desc="Downloading",
                total=total_size,
                unit='B',
                unit_scale=True,
                unit_divisor=1024,
            ) as pbar:
                for chunk in response.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
                        pbar.update(len(chunk))

            print("Download completed! Extracting...")
            with tarfile.open(filename, 'r') as tar:
                tar.extractall(data_dir)

            print(f"Dataset extracted to: {data_dir}/Task01_BrainTumour/")
            return os.path.join(data_dir, 'Task01_BrainTumour')

        except Exception as e:
            print(f"Download failed: {e}")
            print("Please try manual download option.")
            return None

    elif choice == '2':
        print("Academic Torrents option:")
        print(f"Please download the torrent file and extract to: {data_dir}")
        print("Torrent URL: " + download_urls['academic_torrents'])
        return None

    else:
        print("\n=== MANUAL DOWNLOAD INSTRUCTIONS ===")
        print("1. Visit: http://medicaldecathlon.com/dataaws/")
        print("2. Download: Task01_BrainTumour.tar (2GB)")
        print("3. Extract to:", data_dir)
        print("4. The folder structure should be:")
        print("   medical_decathlon_data/")
        print("   └── Task01_BrainTumour/")
        print("       ├── imagesTr/")
        print("       ├── imagesTs/")
        print("       ├── labelsTr/")
        print("       └── dataset.json")
        print("\nAlternative sources:")
        print("- Google Drive: Available on the official website")
        print("- Academic Torrents: https://academictorrents.com/details/274be65156ed14828fb7b30b82407a2417e1924a")

        return input("Enter path to extracted Task01_BrainTumour folder: ").strip()

class MedicalDecathlonBrainDataset(Dataset):
    """Dataset class for Medical Segmentation Decathlon Brain Dataset"""

    def __init__(self, data_root, split='training', transform=None, target_size=(128, 128)):
        self.data_root = data_root
        self.split = split
        self.transform = transform
        self.target_size = target_size

        # Set paths based on split
        if split == 'training':
            self.images_dir = os.path.join(data_root, 'imagesTr')
            self.labels_dir = os.path.join(data_root, 'labelsTr')
        else:
            self.images_dir = os.path.join(data_root, 'imagesTs')
            self.labels_dir = None  # Test set usually doesn't have labels

        # Load dataset information
        self.dataset_info = self.load_dataset_info()

        # Get all image files
        self.image_files = sorted([f for f in os.listdir(self.images_dir)
                                 if f.endswith('.nii.gz')])

        print(f"Found {len(self.image_files)} {split} samples")

    def load_dataset_info(self):
        """Load dataset.json file with metadata"""
        json_path = os.path.join(self.data_root, 'dataset.json')
        if os.path.exists(json_path):
            import json
            with open(json_path, 'r') as f:
                return json.load(f)
        return {}

    def __len__(self):
        return len(self.image_files)

    def load_nifti_volume(self, filepath):
        """Load NIfTI volume and extract modalities"""
        nii_img = nib.load(filepath)
        volume = nii_img.get_fdata()

        # Medical Decathlon brain data has 4 modalities: [T1, T1Gd, T2, FLAIR]
        # Shape is typically (H, W, D, 4) where D is depth (slices)

        if len(volume.shape) == 4:
            # Take middle slice from each modality
            middle_slice = volume.shape[2] // 2

            t1 = volume[:, :, middle_slice, 0]      # T1
            t1gd = volume[:, :, middle_slice, 1]    # T1 with Gadolinium (contrast)
            t2 = volume[:, :, middle_slice, 2]      # T2
            flair = volume[:, :, middle_slice, 3]   # FLAIR

            # For fusion, we'll use T1Gd (contrast-enhanced) as "CT-like" and T2 as "MRI-like"
            ct_like = t1gd  # T1Gd shows enhanced regions (tumor boundaries)
            mri_like = t2   # T2 shows tissue contrast and edema

        else:
            # Fallback if format is different
            if len(volume.shape) == 3:
                middle_slice = volume.shape[2] // 2
                ct_like = mri_like = volume[:, :, middle_slice]
            else:
                ct_like = mri_like = volume

        return ct_like, mri_like

    def load_segmentation_mask(self, image_filename):
        """Load corresponding segmentation mask"""
        if self.labels_dir is None:
            return None

        mask_filename = image_filename  # Same filename for mask
        mask_path = os.path.join(self.labels_dir, mask_filename)

        if os.path.exists(mask_path):
            nii_mask = nib.load(mask_path)
            mask_volume = nii_mask.get_fdata()

            # Take middle slice
            if len(mask_volume.shape) >= 3:
                middle_slice = mask_volume.shape[2] // 2
                mask = mask_volume[:, :, middle_slice]
            else:
                mask = mask_volume

            # Convert to binary (any tumor class > 0)
            binary_mask = (mask > 0).astype(float)
            return binary_mask

        return None

    def preprocess_image(self, img):
        """Preprocess medical image"""
        # Normalize to 0-1 range
        img_min, img_max = img.min(), img.max()
        if img_max > img_min:
            img = (img - img_min) / (img_max - img_min)
        else:
            img = np.zeros_like(img)

        # Resize to target size
        img = cv2.resize(img, self.target_size)

        # Apply CLAHE for contrast enhancement
        img_uint8 = (img * 255).astype(np.uint8)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        img_enhanced = clahe.apply(img_uint8) / 255.0

        return img_enhanced

    def __getitem__(self, idx):
        # Load image
        image_filename = self.image_files[idx]
        image_path = os.path.join(self.images_dir, image_filename)

        # Extract CT-like and MRI-like images
        ct_like, mri_like = self.load_nifti_volume(image_path)

        # Preprocess
        ct_processed = self.preprocess_image(ct_like)
        mri_processed = self.preprocess_image(mri_like)

        # Load segmentation mask for label
        mask = self.load_segmentation_mask(image_filename)

        # Convert to tensors
        ct_tensor = torch.FloatTensor(ct_processed).unsqueeze(0)
        mri_tensor = torch.FloatTensor(mri_processed).unsqueeze(0)

        # For classification: any tumor present = 1, no tumor = 0
        if mask is not None:
            has_tumor = 1 if np.sum(mask) > 0 else 0
            label = torch.LongTensor([has_tumor])
            return ct_tensor, mri_tensor, label
        else:
            return ct_tensor, mri_tensor

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Add these functions and classes after your existing MedicalDecathlonBrainDataset class

def clean_dataset_path(dataset_path):
    """Remove hidden system files and validate dataset"""
    if not os.path.exists(dataset_path):
        logger.error(f"Dataset path does not exist: {dataset_path}")
        return False

    # Remove macOS hidden files
    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.startswith('._') or file == '.DS_Store':
                hidden_file_path = os.path.join(root, file)
                try:
                    os.remove(hidden_file_path)
                    logger.info(f"Removed hidden file: {hidden_file_path}")
                except Exception as e:
                    logger.warning(f"Could not remove {hidden_file_path}: {e}")

    return True

def validate_nifti_file(filepath):
    """Validate if a file is a proper NIfTI file"""
    try:
        # Check file extension
        if not (filepath.endswith('.nii') or filepath.endswith('.nii.gz')):
            return False

        # Check if file exists and is not empty
        if not os.path.exists(filepath) or os.path.getsize(filepath) == 0:
            return False

        # Try to load the file
        nii_img = nib.load(filepath)
        data = nii_img.get_fdata()

        # Check if data is valid
        if data.size == 0 or np.all(np.isnan(data)):
            return False

        return True
    except Exception as e:
        logger.warning(f"File validation failed for {filepath}: {e}")
        return False

class RobustMedicalDecathlonDataset(Dataset):
    """Robust Medical Decathlon Dataset with file validation - USE THIS INSTEAD"""

    def __init__(self, data_dir, split='train', transform=None, max_samples=None):
        self.data_dir = data_dir
        self.split = split
        self.transform = transform

        # Clean dataset first
        logger.info("Cleaning dataset and removing system files...")
        clean_dataset_path(data_dir)

        # Get image and label paths
        if split == 'train':
            self.image_dir = os.path.join(data_dir, 'imagesTr')
            self.label_dir = os.path.join(data_dir, 'labelsTr')
        else:
            self.image_dir = os.path.join(data_dir, 'imagesTs')
            self.label_dir = os.path.join(data_dir, 'labelsTs')

        # Get valid image files
        self.image_files = self._get_valid_files()

        if max_samples:
            self.image_files = self.image_files[:max_samples]

        logger.info(f"Found {len(self.image_files)} valid files for {split} split")

        if len(self.image_files) == 0:
            raise ValueError(f"No valid NIfTI files found in {self.image_dir}")

    def _get_valid_files(self):
        """Get list of valid NIfTI files"""
        valid_files = []

        # Look for .nii.gz and .nii files
        pattern_gz = os.path.join(self.image_dir, "*.nii.gz")
        pattern_nii = os.path.join(self.image_dir, "*.nii")

        all_files = glob.glob(pattern_gz) + glob.glob(pattern_nii)

        logger.info(f"Found {len(all_files)} potential files")

        for filepath in sorted(all_files):
            # Skip hidden files
            filename = os.path.basename(filepath)
            if filename.startswith('._') or filename.startswith('.'):
                logger.info(f"Skipping hidden file: {filename}")
                continue

            # Validate file
            if validate_nifti_file(filepath):
                valid_files.append(filepath)
                logger.debug(f"Valid file: {filename}")
            else:
                logger.warning(f"Invalid file: {filename}")

        return valid_files

    def load_nifti_volume(self, filepath):
        """Load NIfTI volume and extract modalities with error handling"""
        try:
            nii_img = nib.load(filepath)
            volume = nii_img.get_fdata()

            # Handle different volume shapes
            if len(volume.shape) == 4:
                # Multi-modal volume (e.g., FLAIR, T1, T1ce, T2)
                if volume.shape[-1] >= 2:
                    ct_like = volume[:, :, :, 0]  # First modality as CT-like
                    mri_like = volume[:, :, :, 1]  # Second modality as MRI-like
                else:
                    # Single modality, duplicate for both
                    ct_like = volume[:, :, :, 0]
                    mri_like = volume[:, :, :, 0]
            elif len(volume.shape) == 3:
                # Single 3D volume, use same for both modalities
                ct_like = volume
                mri_like = volume
            else:
                raise ValueError(f"Unexpected volume shape: {volume.shape}")

            # Normalize volumes
            ct_like = self.normalize_volume(ct_like)
            mri_like = self.normalize_volume(mri_like)

            # Select middle slice for 2D processing
            mid_slice = ct_like.shape[2] // 2
            ct_slice = ct_like[:, :, mid_slice]
            mri_slice = mri_like[:, :, mid_slice]

            # Resize to standard size (256x256)
            ct_slice = self.resize_slice(ct_slice, (256, 256))
            mri_slice = self.resize_slice(mri_slice, (256, 256))

            return ct_slice, mri_slice

        except Exception as e:
            logger.error(f"Error loading {filepath}: {e}")
            # Return dummy data to prevent crash
            dummy_slice = np.zeros((256, 256), dtype=np.float32)
            return dummy_slice, dummy_slice

    def normalize_volume(self, volume):
        """Normalize volume to [0, 1] range"""
        volume = volume.astype(np.float32)

        # Remove NaN and inf values
        volume = np.nan_to_num(volume, nan=0.0, posinf=1.0, neginf=0.0)

        # Normalize to [0, 1]
        vol_min, vol_max = volume.min(), volume.max()
        if vol_max > vol_min:
            volume = (volume - vol_min) / (vol_max - vol_min)

        return volume

    def resize_slice(self, slice_2d, target_size):
        """Resize 2D slice to target size using interpolation"""
        current_size = slice_2d.shape
        zoom_factors = (target_size[0] / current_size[0], target_size[1] / current_size[1])

        resized_slice = zoom(slice_2d, zoom_factors, order=1)  # Linear interpolation
        return resized_slice

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        image_path = self.image_files[idx]

        # Extract CT-like and MRI-like images
        ct_like, mri_like = self.load_nifti_volume(image_path)

        # Convert to tensors and add channel dimension
        ct_tensor = torch.from_numpy(ct_like).unsqueeze(0).float()  # Shape: [1, H, W]
        mri_tensor = torch.from_numpy(mri_like).unsqueeze(0).float()  # Shape: [1, H, W]

        # Apply transforms if provided
        if self.transform:
            ct_tensor = self.transform(ct_tensor)
            mri_tensor = self.transform(mri_tensor)

        # For simplicity, create a dummy label (0 for no tumor, 1 for tumor)
        # In real implementation, you would load actual labels
        label = torch.randint(0, 2, (1,)).float()

        return ct_tensor, mri_tensor, label

# Updated fusion dataset wrapper
class FusionDataset(Dataset):
    """Wrapper for fusion training (no labels needed)"""
    def __init__(self, base_dataset):
        self.base_dataset = base_dataset

    def __len__(self):
        return len(self.base_dataset)

    def __getitem__(self, idx):
        try:
            ct, mri, _ = self.base_dataset[idx]  # Ignore labels for fusion training
            return ct, mri
        except Exception as e:
            logger.error(f"Error fetching item {idx}: {e}")
            # Return dummy data to prevent crash
            dummy_tensor = torch.zeros(1, 256, 256)
            return dummy_tensor, dummy_tensor

def create_robust_dataloaders(data_dir, batch_size=4, num_workers=2, max_samples=50):
    """Create robust dataloaders with error handling"""

    try:
        # Create datasets using the robust class
        train_dataset = RobustMedicalDecathlonDataset(
            data_dir,
            split='train',
            max_samples=max_samples
        )

        # Create fusion dataset (no labels)
        fusion_dataset = FusionDataset(train_dataset)

        # Create dataloaders
        fusion_loader = DataLoader(
            fusion_dataset,
            batch_size=batch_size,
            shuffle=True,
            num_workers=num_workers,
            drop_last=True,  # Drop incomplete batches
            pin_memory=torch.cuda.is_available()
        )

        # Create validation loader (using subset of training data)
        val_size = min(20, len(train_dataset) // 4)
        val_indices = list(range(0, val_size))
        val_subset = torch.utils.data.Subset(train_dataset, val_indices)

        val_loader = DataLoader(
            val_subset,
            batch_size=batch_size,
            shuffle=False,
            num_workers=num_workers,
            drop_last=True,
            pin_memory=torch.cuda.is_available()
        )

        logger.info(f"Created dataloaders - Fusion: {len(fusion_loader)} batches, Val: {len(val_loader)} batches")

        return fusion_loader, val_loader, train_dataset

    except Exception as e:
        logger.error(f"Error creating dataloaders: {e}")
        raise

def test_dataset_loading(data_dir):
    """Test dataset loading with error handling"""
    try:
        logger.info("Testing dataset loading...")

        # Clean and test dataset
        if not clean_dataset_path(data_dir):
            logger.error("Dataset path validation failed")
            return False

        # Try creating dataset
        dataset = RobustMedicalDecathlonDataset(data_dir, max_samples=5)

        if len(dataset) == 0:
            logger.error("No valid samples found in dataset")
            return False

        # Test loading a sample
        ct, mri, label = dataset[0]
        logger.info(f"Sample loaded successfully - CT: {ct.shape}, MRI: {mri.shape}, Label: {label}")

        # Test dataloader
        fusion_loader, val_loader, _ = create_robust_dataloaders(data_dir, batch_size=2, max_samples=5)

        # Test one batch
        for batch_idx, (ct_batch, mri_batch) in enumerate(fusion_loader):
            logger.info(f"Batch {batch_idx} loaded - CT: {ct_batch.shape}, MRI: {mri_batch.shape}")
            break

        logger.info("Dataset loading test passed!")
        return True

    except Exception as e:
        logger.error(f"Dataset loading test failed: {e}")
        return False

def setup_medical_decathlon_dataset(data_root):
    """Setup Medical Decathlon dataset for training"""

    # Verify dataset structure
    required_dirs = ['imagesTr', 'labelsTr', 'dataset.json']
    missing_dirs = []

    for req_dir in required_dirs:
        if not os.path.exists(os.path.join(data_root, req_dir)):
            missing_dirs.append(req_dir)

    if missing_dirs:
        print(f"ERROR: Missing required directories/files: {missing_dirs}")
        print(f"Please ensure the dataset is properly extracted at: {data_root}")
        return None, None, None

    print("✅ Dataset structure verified!")

    # Load dataset info
    with open(os.path.join(data_root, 'dataset.json'), 'r') as f:
        dataset_info = json.load(f)

    print(f"Dataset: {dataset_info.get('name', 'Unknown')}")
    print(f"Description: {dataset_info.get('description', 'No description')}")
    print(f"Modalities: {dataset_info.get('modality', 'Unknown')}")
    print(f"Labels: {dataset_info.get('labels', 'Unknown')}")

    # Create datasets
    full_dataset = MedicalDecathlonBrainDataset(data_root, split='training')

    # Split into train/validation
    train_size = int(0.8 * len(full_dataset))
    val_size = len(full_dataset) - train_size

    train_dataset, val_dataset = torch.utils.data.random_split(
        full_dataset, [train_size, val_size]
    )

    # For fusion training (unsupervised)
    fusion_dataset = MedicalDecathlonBrainDataset(data_root, split='training')

    print(f"Training samples: {len(train_dataset)}")
    print(f"Validation samples: {len(val_dataset)}")
    print(f"Total fusion samples: {len(fusion_dataset)}")

    return train_dataset, val_dataset, fusion_dataset

# ==================== ENHANCED VISUALIZATION FUNCTIONS ====================

def visualize_detailed_metastases_detection(ct_img, mri_img, fused_img, detections,
                                          case_info=None, save_path=None):
    """
    Enhanced visualization for metastases detection with detailed annotations

    Parameters:
    -----------
    ct_img: numpy array
        Original CT image
    mri_img: numpy array
        Original MRI image
    fused_img: numpy array
        Fused CT-MRI image
    detections: list
        List of detected metastases regions
    case_info: dict, optional
        Patient and case information
    save_path: str, optional
        Path to save the visualization
    """
    # Create a comprehensive figure
    fig = plt.figure(figsize=(20, 15))
    gs = gridspec.GridSpec(3, 3, height_ratios=[1, 1, 0.5])

    # Plot original images
    ax1 = plt.subplot(gs[0, 0])
    ax1.imshow(ct_img, cmap='gray')
    ax1.set_title('CT Image', fontsize=12, fontweight='bold')
    ax1.axis('off')

    ax2 = plt.subplot(gs[0, 1])
    ax2.imshow(mri_img, cmap='gray')
    ax2.set_title('MRI Image', fontsize=12, fontweight='bold')
    ax2.axis('off')

    ax3 = plt.subplot(gs[0, 2])
    ax3.imshow(fed_img, cmap='gray')
    ax3.set_title('Fused Image', fontsize=12, fontweight='bold')
    ax3.axis('off')

    # Plot enhanced views with detections
    ax4 = plt.subplot(gs[1, 0])
    ax4.imshow(ct_img, cmap='hot')
    ax4.set_title('CT (Enhanced View)', fontsize=12, fontweight='bold')
    ax4.axis('off')

    ax5 = plt.subplot(gs[1, 1])
    ax5.imshow(mri_img, cmap='viridis')
    ax5.set_title('MRI (Enhanced View)', fontsize=12, fontweight='bold')
    ax5.axis('off')

    # Plot fused image with detections
    ax6 = plt.subplot(gs[1, 2])
    ax6.imshow(fused_img, cmap='gray')

    # Draw bounding boxes for detections
    for i, detection in enumerate(detections):
        x, y, w, h = detection['bbox']
        confidence = detection['confidence']

        # Draw rectangle
        rect = patches.Rectangle((x, y), w, h, linewidth=2,
                               edgecolor='red', facecolor='none')
        ax6.add_patch(rect)

        # Add text with confidence
        ax6.text(x, y-5, f'Metastasis {i+1}\n({confidence:.2f})',
                color='red', fontsize=10, fontweight='bold',
                bbox=dict(facecolor='white', alpha=0.7, edgecolor='red'))

    ax6.set_title('Detected Metastases', fontsize=12, fontweight='bold')
    ax6.axis('off')

    # Create heatmap of suspicious regions
    ax7 = plt.subplot(gs[2, :])
    heatmap = np.zeros_like(fused_img)

    for detection in detections:
        x, y, w, h = detection['bbox']
        confidence = detection['confidence']

        # Create a Gaussian blob at detection location
        center_x, center_y = x + w//2, y + h//2

        # Create a meshgrid
        y_grid, x_grid = np.ogrid[:fused_img.shape[0], :fused_img.shape[1]]

        # Calculate distance from center
        distance = np.sqrt((x_grid - center_x)**2 + (y_grid - center_y)**2)

        # Create Gaussian (higher confidence = more intense)
        gaussian = np.exp(-(distance**2) / (2 * (20 * confidence)**2))
        heatmap = np.maximum(heatmap, gaussian * confidence)

    # Plot heatmap
    im = ax7.imshow(heatmap, cmap='jet', alpha=0.7)
    ax7.imshow(fused_img, cmap='gray', alpha=0.3)
    ax7.set_title('Suspicious Regions Heatmap', fontsize=12, fontweight='bold')
    ax7.axis('off')

    # Add colorbar
    cbar = plt.colorbar(im, ax=ax7, fraction=0.046, pad=0.04)
    cbar.set_label('Detection Confidence', fontsize=10)

    # Add case information if provided
    if case_info:
        plt.figtext(0.02, 0.02,
                   f"Case ID: {case_info.get('case_id', 'N/A')}\n"
                   f"Patient Age: {case_info.get('age', 'N/A')}\n"
                   f"Detection Date: {case_info.get('date', 'N/A')}\n"
                   f"Number of Detections: {len(detections)}",
                   bbox=dict(facecolor='white', alpha=0.8),
                   fontsize=10)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Visualization saved to: {save_path}")

    plt.show()

def generate_clinical_report(analysis_results, patient_info=None):
    """
    Generate a comprehensive clinical report based on analysis results

    Parameters:
    -----------
    analysis_results: dict
        Results from the metastases analysis
    patient_info: dict, optional
        Patient information

    Returns:
    --------
    report: str
        Formatted clinical report
    """
    report = f"""
    BRAIN METASTASES DETECTION REPORT
    =================================

    Patient Information:
    -------------------
    ID: {patient_info.get('id', 'N/A') if patient_info else 'N/A'}
    Age: {patient_info.get('age', 'N/A') if patient_info else 'N/A'}
    Sex: {patient_info.get('sex', 'N/A') if patient_info else 'N/A'}
    Study Date: {patient_info.get('date', 'N/A') if patient_info else 'N/A'}

    Analysis Results:
    ----------------
    Metastases Detected: {'YES' if analysis_results['has_metastases'] else 'NO'}
    Confidence Level: {analysis_results['confidence']:.2%}
    Number of Detected Regions: {analysis_results['num_detected_regions']}

    Detected Regions:
    -----------------
    """

    for i, detection in enumerate(analysis_results['detected_regions']):
        x, y, w, h = detection['bbox']
        report += f"""
    Region {i+1}:
      - Location: (x={x}, y={y})
      - Size: {w}×{h} pixels
      - Area: {detection['area']} pixels²
      - Confidence: {detection['confidence']:.3f}
    """

    report += f"""
    Radiomic Features:
    -----------------
    Mean Intensity: {analysis_results['radiomic_features']['mean']:.4f}
    Standard Deviation: {analysis_results['radiomic_features']['std']:.4f}
    Entropy: {analysis_results['radiomic_features']['entropy']:.4f}
    Contrast: {analysis_results['radiomic_features']['contrast']:.4f}

    Fusion Quality Metrics:
    ----------------------
    SSIM: {analysis_results['fusion_quality']['ssim']:.4f}
    PSNR: {analysis_results['fusion_quality']['psnr']:.2f} dB
    Edge Preservation: {analysis_results['fusion_quality']['edge_preservation']:.4f}

    Clinical Recommendations:
    ------------------------
    """

    if analysis_results['has_metastases']:
        report += """
    - Urgent radiologist review recommended
    - Consider contrast-enhanced MRI for confirmation
    - Oncology consultation advised
    - Treatment planning required based on number and location of metastases
    """
    else:
        report += """
    - No evidence of brain metastases
    - Routine follow-up as per clinical guidelines
    - Consider repeat imaging if clinical suspicion persists
    """

    report += """
    Disclaimer:
    -----------
    This report is generated by an AI system and should be reviewed by a qualified radiologist.
    The findings are not a substitute for professional medical judgment.
    """

    return report

def interactive_visualization(fusion_model, classifier, dataset, case_index=0):
    """
    Interactive visualization of detection results

    Parameters:
    -----------
    fusion_model: EnhancedFusionAutoencoder
        Trained fusion model
    classifier: MetastasesClassifier
        Trained classifier model
    dataset: Dataset
        Medical dataset
    case_index: int
        Index of the case to visualize
    """
    # Get the case data
    ct_tensor, mri_tensor, label = dataset[case_index]

    # Move to device
    ct_tensor = ct_tensor.unsqueeze(0).to(device)
    mri_tensor = mri_tensor.unsqueeze(0).to(device)

    # Set models to evaluation mode
    fusion_model.eval()
    classifier.eval()

    with torch.no_grad():
        # Generate fused image
        fused_img = fusion_model(ct_tensor, mri_tensor)

        # Classify
        fused_3ch = fused_img.repeat(1, 3, 1, 1)
        outputs = classifier(fused_3ch)
        probabilities = torch.softmax(outputs, dim=1)

        # Convert to numpy
        ct_np = ct_tensor[0, 0].cpu().numpy()
        mri_np = mri_tensor[0, 0].cpu().numpy()
        fused_np = fused_img[0, 0].cpu().numpy()

        # Detect metastases regions
        detections = detect_metastases_regions(fused_np)

        # Create analysis results
        analysis_results = {
            'has_metastases': torch.argmax(probabilities).item() == 1,
            'confidence': torch.max(probabilities).item(),
            'num_detected_regions': len(detections),
            'detected_regions': detections,
            'radiomic_features': extract_radiomic_features(fused_np),
            'fusion_quality': evaluate_fusion_quality(fused_np, ct_np, mri_np)
        }

    # Generate visualization
    case_info = {
        'case_id': f'Case_{case_index:04d}',
        'date': pd.Timestamp.now().strftime('%Y-%m-%d')
    }

    visualize_detailed_metastases_detection(
        ct_np, mri_np, fused_np, detections, case_info
    )

    # Generate and print clinical report
    report = generate_clinical_report(analysis_results)
    print(report)

    return analysis_results

def batch_visualization(fusion_model, classifier, dataset, num_cases=5):
    """
    Visualize multiple cases in a batch

    Parameters:
    -----------
    fusion_model: EnhancedFusionAutoencoder
        Trained fusion model
    classifier: MetastasesClassifier
        Trained classifier model
    dataset: Dataset
        Medical dataset
    num_cases: int
        Number of cases to visualize
    """
    # Select random cases
    indices = np.random.choice(len(dataset), min(num_cases, len(dataset)), replace=False)

    all_results = []

    for i, idx in enumerate(indices):
        print(f"\n{'='*50}")
        print(f"Visualizing Case {i+1}/{num_cases} (Index: {idx})")
        print(f"{'='*50}")

        results = interactive_visualization(fusion_model, classifier, dataset, idx)
        all_results.append(results)

        # Add a pause between cases
        if i < num_cases - 1:
            input("\nPress Enter to continue to next case...")

    return all_results

# ==================== UPDATED MAIN FUNCTION ====================

def main_with_real_data():
    """Main function using real Medical Decathlon data"""
    print("=== Brain Metastases Detection - Real Medical Decathlon Data ===")
    print("Using Task01_BrainTumour from Medical Segmentation Decathlon")

    # Step 1: Download/setup dataset
    print("\nStep 1: Setting up Medical Decathlon dataset...")

    # Check if dataset exists
    data_root = './medical_decathlon_data/Task01_BrainTumour'

    if not os.path.exists(data_root):
        print("Medical Decathlon dataset not found. Starting download process...")
        data_root = download_medical_decathlon_brain_data()

        if data_root is None:
            print("Please download the dataset manually and rerun.")
            return
    else:
        print(f"Dataset found at: {data_root}")

    # Clean dataset before loading
    print("Cleaning dataset directory...")
    if not clean_dataset_path(data_root):
        print("Warning: Dataset cleaning failed. Proceeding with potential file issues.")

    # Setup datasets
    train_dataset, val_dataset, fusion_dataset = setup_medical_decathlon_dataset(data_root)

    if train_dataset is None:
        print("Failed to setup dataset. Please check the data structure.")
        return

    # Step 2: Create data loaders
    print("\nStep 2: Creating data loaders...")

    # Create a fusion-only dataset that returns just CT and MRI
    class FusionOnlyDataset(torch.utils.data.Dataset):
        def __init__(self, base_dataset):
            self.base_dataset = base_dataset

        def __len__(self):
            return len(self.base_dataset)

        def __getitem__(self, idx):
            ct, mri, _ = self.base_dataset[idx]  # Ignore labels for fusion training
            return ct, mri

    fusion_only_dataset = FusionOnlyDataset(fusion_dataset)
    fusion_loader = DataLoader(fusion_only_dataset, batch_size=1, shuffle=True)
    train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)  # Smaller batch due to memory
    val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False)

    # Step 3: Train fusion model
    print("\nStep 3: Training fusion model on real data...")
    fusion_model = EnhancedFusionAutoencoder().to(device)

    # Train with real medical data
    fusion_history = train_fusion_model(fusion_model, fusion_loader, num_epochs=20)

    # Step 4: Train classification model
    print("\nStep 4: Training classification model on real data...")
    classifier, classification_history = train_classification_model(fusion_model, train_loader, num_epochs=20)

    # Step 5: Evaluate on real data
    print("\nStep 5: Evaluating on real medical data...")
    evaluate_real_medical_data(fusion_model, classifier, val_loader)

    # Step 6: Clinical demonstration
    print("\nStep 6: Clinical demonstration with real data...")
    clinical_demo_real_data(fusion_model, classifier, val_loader, data_root)

    # Step 7: Enhanced visualization
    print("\nStep 7: Enhanced visualization of results...")
    enhanced_visualization_demo(fusion_model, classifier, val_loader)

    return fusion_model, classifier, val_loader

def enhanced_visualization_demo(fusion_model, classifier, val_loader):
    """Enhanced visualization demonstration"""
    print("\n=== ENHANCED VISUALIZATION DEMONSTRATION ===")

    # Get a batch of data
    for ct_batch, mri_batch, labels in val_loader:
        # Use the first sample in the batch
        ct_tensor = ct_batch[0:1].to(device)
        mri_tensor = mri_batch[0:1].to(device)
        label = labels[0]

        # Set models to evaluation mode
        fusion_model.eval()
        classifier.eval()

        with torch.no_grad():
            # Generate fused image
            fused_img = fusion_model(ct_tensor, mri_tensor)

            # Classify
            fused_3ch = fused_img.repeat(1, 3, 1, 1)
            outputs = classifier(fused_3ch)
            probabilities = torch.softmax(outputs, dim=1)

            # Convert to numpy
            ct_np = ct_tensor[0, 0].cpu().numpy()
            mri_np = mri_tensor[0, 0].cpu().numpy()
            fused_np = fused_img[0, 0].cpu().numpy()

            # Detect metastases regions
            detections = detect_metastases_regions(fused_np)

            # Create analysis results
            analysis_results = {
                'has_metastases': torch.argmax(probabilities).item() == 1,
                'confidence': torch.max(probabilities).item(),
                'num_detected_regions': len(detections),
                'detected_regions': detections,
                'radiomic_features': extract_radiomic_features(fused_np),
                'fusion_quality': evaluate_fusion_quality(fused_np, ct_np, mri_np)
            }

        # Generate visualization
        case_info = {
            'case_id': 'Demo_Case_001',
            'age': '52',
            'sex': 'M',
            'date': pd.Timestamp.now().strftime('%Y-%m-%d')
        }

        print(f"\nGround Truth: {'Metastases Present' if label.item() == 1 else 'No Metastases'}")
        print(f"AI Prediction: {'Metastases Present' if analysis_results['has_metastases'] else 'No Metastases'}")
        print(f"Confidence: {analysis_results['confidence']:.3f}")
        print(f"Number of Detected Regions: {len(detections)}")

        visualize_detailed_metastases_detection(
            ct_np, mri_np, fused_np, detections, case_info
        )

        # Generate and print clinical report
        report = generate_clinical_report(analysis_results, case_info)
        print(report)

        break

def evaluate_real_medical_data(fusion_model, classifier, val_loader):
    """Evaluate models on real medical data"""
    fusion_model.eval()
    classifier.eval()

    all_predictions = []
    all_labels = []
    fusion_metrics = []

    print("Evaluating on real medical validation data...")

    with torch.no_grad():
        for batch_idx, (ct_batch, mri_batch, labels) in enumerate(val_loader):
            ct_batch, mri_batch = ct_batch.to(device), mri_batch.to(device)
            labels = labels.squeeze()

            # Generate fused images
            fused_imgs = fusion_model(ct_batch, mri_batch)

            # Evaluate fusion quality
            for i in range(ct_batch.size(0)):
                ct_np = ct_batch[i, 0].cpu().numpy()
                mri_np = mri_batch[i, 0].cpu().numpy()
                fused_np = fused_imgs[i, 0].cpu().numpy()

                metrics = evaluate_fusion_quality(fused_np, ct_np, mri_np)
                fusion_metrics.append(metrics)

            # Classification
            fused_imgs_3ch = fused_imgs.repeat(1, 3, 1, 1)
            outputs = classifier(fused_imgs_3ch)
            _, predicted = torch.max(outputs, 1)

            all_predictions.extend(predicted.cpu().numpy().flatten())
            all_labels.extend(labels.cpu().numpy().flatten())

            if batch_idx == 0:  # Show first batch results
                print(f"Sample results - True: {labels.numpy()}, Predicted: {predicted.cpu().numpy()}")

    # Calculate metrics
    avg_fusion_metrics = {
        'ssim': np.mean([m['ssim'] for m in fusion_metrics]),
        'psnr': np.mean([m['psnr'] for m in fusion_metrics]),
        'entropy': np.mean([m['entropy'] for m in fusion_metrics]),
        'edge_preservation': np.mean([m['edge_preservation'] for m in fusion_metrics])
    }

    print("\n=== REAL DATA FUSION QUALITY METRICS ===")
    print(f"Average SSIM: {avg_fusion_metrics['ssim']:.4f}")
    print(f"Average PSNR: {avg_fusion_metrics['psnr']:.2f} dB")
    print(f"Average Entropy: {avg_fusion_metrics['entropy']:.4f}")
    print(f"Edge Preservation: {avg_fusion_metrics['edge_preservation']:.4f}")

    # Classification metrics
    unique_labels = np.unique(all_labels)
    print(f"\n=== REAL DATA CLASSIFICATION METRICS ===")
    print(f"Unique labels: {unique_labels}")
    print(f"Total samples: {len(all_labels)}")
    print(f"Tumor cases: {np.sum(all_labels)}")
    print(f"Normal cases: {len(all_labels) - np.sum(all_labels)}")

    if len(unique_labels) > 1:
        print("\nClassification Report:")
        print(classification_report(all_labels, all_predictions,
                                  target_names=['No Tumor', 'Tumor Present']))

        # Confusion Matrix
        cm = confusion_matrix(all_labels, all_predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=['No Tumor', 'Tumor Present'],
                    yticklabels=['No Tumor', 'Tumor Present'])
        plt.title('Confusion Matrix - Real Brain Tumor Data')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()
    else:
        accuracy = np.mean(np.array(all_predictions) == np.array(all_labels))
        print(f"Accuracy: {accuracy * 100:.2f}%")

def clinical_demo_real_data(fusion_model, classifier, val_loader, data_root):
    """Clinical demonstration with real medical data"""
    print("\n=== CLINICAL DEMONSTRATION WITH REAL DATA ===")

    fusion_model.eval()
    classifier.eval()

    demo_count = 0
    with torch.no_grad():
        for ct_batch, mri_batch, labels in val_loader:
            if demo_count >= 3:
                break

            ct_batch, mri_batch = ct_batch.to(device), mri_batch.to(device)
            labels = labels.squeeze()

            # Generate fused image
            fused_img = fusion_model(ct_batch, mri_batch)

            # Classification
            fused_3ch = fused_img.repeat(1, 3, 1, 1)
            outputs = classifier(fused_3ch)
            probabilities = torch.softmax(outputs, dim=1)

            # Process first sample in batch
            ct_np = ct_batch[0, 0].cpu().numpy()
            mri_np = mri_batch[0, 0].cpu().numpy()
            fused_np = fused_img[0, 0].cpu().numpy()

            print(f"\n--- Real Medical Sample {demo_count + 1} ---")
            print(f"Ground Truth: {'Tumor Present' if labels[0] == 1 else 'No Tumor'}")
            print(f"AI Prediction: {'Tumor Present' if torch.argmax(probabilities[0]) == 1 else 'No Tumor'}")
            print(f"Confidence: {torch.max(probabilities[0]):.3f}")

            # Detect potential tumor regions
            detections = detect_metastases_regions(fused_np, threshold=0.7)
            print(f"Detected suspicious regions: {len(detections)}")

            # Enhanced visualization for medical data
            visualize_medical_fusion_results(ct_np, mri_np, fused_np, detections,
                                           title_suffix=f"Real Medical Sample {demo_count + 1}")

            demo_count += 1

def visualize_medical_fusion_results(ct_img, mri_img, fused_img, detections=None, title_suffix=""):
    """Enhanced visualization for medical data"""
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Original modalities
    axes[0, 0].imshow(ct_img, cmap='gray')
    axes[0, 0].set_title(f'T1Gd (Contrast Enhanced) {title_suffix}')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(mri_img, cmap='gray')
    axes[0, 1].set_title(f'T2 Weighted {title_suffix}')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(fused_img, cmap='gray')
    axes[0, 2].set_title(f'AI Fused Image {title_suffix}')
    axes[0, 2].axis('off')

    # Enhanced views
    axes[1, 0].imshow(ct_img, cmap='hot')
    axes[1, 0].set_title('T1Gd (Hot Colormap)')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(mri_img, cmap='viridis')
    axes[1, 1].set_title('T2 (Enhanced View)')
    axes[1, 1].axis('off')

    # Fused with AI detections
    axes[1, 2].imshow(fused_img, cmap='gray')
    if detections:
        for i, detection in enumerate(detections):
            x, y, w, h = detection['bbox']
            rect = plt.Rectangle((x, y), w, h, linewidth=2,
                               edgecolor='red', facecolor='none')
            axes[1, 2].add_patch(rect)
            axes[1, 2].text(x, y-2, f'ROI-{i+1}', color='red', fontsize=8)
    axes[1, 2].set_title('AI Analysis + ROI Detection')
    axes[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

# Import additional required libraries

def preprocess_brain_images(image_path, target_size=(128, 128)):
    """Enhanced preprocessing for brain images"""
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply CLAHE for contrast enhancement
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)

    # Gaussian smoothing to reduce noise
    img = cv2.GaussianBlur(img, (3, 3), 0)

    # Resize
    img = cv2.resize(img, target_size)

    # Normalize
    img = img / 255.0

    return img

def create_synthetic_metastases_data(base_dir, num_samples=1000):
    """Create synthetic training data with simulated metastases"""
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)
        os.makedirs(os.path.join(base_dir, 'ct'))
        os.makedirs(os.path.join(base_dir, 'mri'))
        os.makedirs(os.path.join(base_dir, 'labels'))

    labels_data = []

    # Ensure balanced dataset
    num_positive = int(num_samples * 0.4)  # 40% positive cases
    num_negative = num_samples - num_positive

    print(f"Creating {num_positive} metastases cases and {num_negative} normal cases...")

    for i in range(num_samples):
        # Create synthetic brain-like images
        ct_img = create_synthetic_brain_ct()
        mri_img = create_synthetic_brain_mri()

        # First create positive cases, then negative
        has_metastases = i < num_positive

        if has_metastases:
            ct_img, mri_img = add_synthetic_metastases(ct_img, mri_img)
            label = 1
        else:
            label = 0

        # Save images
        ct_filename = f'ct_{i:04d}.png'
        mri_filename = f'mri_{i:04d}.png'

        cv2.imwrite(os.path.join(base_dir, 'ct', ct_filename), (ct_img * 255).astype(np.uint8))
        cv2.imwrite(os.path.join(base_dir, 'mri', mri_filename), (mri_img * 255).ast(np.uint8))

        labels_data.append({'filename': ct_filename, 'has_metastases': label})

    # Save labels
    labels_df = pd.DataFrame(labels_data)
    labels_df.to_csv(os.path.join(base_dir, 'labels', 'labels.csv'), index=False)

    print(f"Created {num_samples} synthetic samples in {base_dir}")
    print(f"Positive cases: {labels_df['has_metastases'].sum()}")
    print(f"Negative cases: {len(labels_df) - labels_df['has_metastases'].sum()}")
    return labels_df

def create_synthetic_brain_ct():
    """Create synthetic CT brain image"""
    img = np.zeros((128, 128), dtype=np.float32)

    # Create brain outline (circular)
    center = (64, 64)
    radius = 50
    y, x = np.ogrid[:128, :128]
    mask = (x - center[0])**2 + (y - center[1])**2 <= radius**2
    img[mask] = 0.7

    # Add skull (higher intensity)
    skull_mask = ((x - center[0])**2 + (y - center[1])**2 <= radius**2) & \
                 ((x - center[0])**2 + (y - center[1])**2 >= (radius-5)**2)
    img[skull_mask] = 0.9

    # Add some internal structures
    img[55:75, 55:75] = 0.5  # Ventricles
    img[45:85, 45:85] += np.random.normal(0, 0.1, (40, 40))  # Texture

    # Add noise
    img += np.random.normal(0, 0.05, img.shape)
    img = np.clip(img, 0, 1)

    return img

def create_synthetic_brain_mri():
    """Create synthetic MRI brain image"""
    img = np.zeros((128, 128), dtype=np.float32)

    # Create brain outline
    center = (64, 64)
    radius = 50
    y, x = np.ogrid[:128, :128]
    mask = (x - center[0])**2 + (y - center[1])**2 <= radius**2
    img[mask] = 0.6

    # Add different tissue contrasts
    # Gray matter
    gray_matter = ((x - center[0])**2 + (y - center[1])**2 <= (radius-10)**2)
    img[gray_matter] = 0.8

    # White matter
    white_matter = ((x - center[0])**2 + (y - center[1])**2 <= (radius-20)**2)
    img[white_matter] = 0.9

    # CSF spaces
    img[60:68, 60:68] = 0.3

    # Add texture
    img[30:98, 30:98] += np.random.normal(0, 0.08, (68, 68))

    # Add noise
    img += np.random.normal(0, 0.03, img.shape)
    img = np.clip(img, 0, 1)

    return img

def add_synthetic_metastases(ct_img, mri_img, num_lesions=None):
    """Add synthetic metastases to brain images"""
    if num_lesions is None:
        num_lesions = np.random.randint(1, 4)  # 1-3 lesions

    for _ in range(num_lesions):
        # Random location within brain area
        center_x = np.random.randint(20, 108)
        center_y = np.random.randint(20, 108)

        # Random size (2-8 pixels radius)
        radius = np.random.randint(2, 8)

        # Create circular lesion
        y, x = np.ogrid[:128, :128]
        lesion_mask = (x - center_x)**2 + (y - center_y)**2 <= radius**2

        # Different appearance in CT vs MRI
        ct_img[lesion_mask] = np.random.uniform(0.8, 1.0)  # Hyperdense in CT
        mri_img[lesion_mask] = np.random.uniform(0.9, 1.0)  # Hyperintense in MRI

        # Add surrounding edema in MRI
        edema_mask = ((x - center_x)**2 + (y - center_y)**2 <= (radius+3)**2) & \
                     ((x - center_x)**2 + (y - center_y)**2 > radius**2)
        mri_img[edema_mask] *= 1.2  # Increased signal around lesion

    return ct_img, mri_img

# ==================== TRAINING FUNCTIONS ====================

def train_fusion_model(model, dataloader, num_epochs=15, lr=0.001):
    """Train the fusion model"""
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)

    model.train()
    training_history = {'loss': [], 'ssim': [], 'psnr': []}

    for epoch in range(num_epochs):
        epoch_loss = 0.0
        epoch_ssim = 0.0
        epoch_psnr = 0.0
        num_batches = 0

        for batch_idx, (ct_batch, mri_batch) in enumerate(dataloader):
            ct_batch, mri_batch = ct_batch.to(device), mri_batch.to(device)

            # Ground truth: enhanced average (you can modify this)
            ground_truth = (ct_batch * 0.6 + mri_batch * 0.4)  # Weighted average

            # Forward pass
            optimizer.zero_grad()
            fused_output = model(ct_batch, mri_batch)

            # Calculate loss
            loss = criterion(fused_output, ground_truth)

            # Backward pass
            loss.backward()
            optimizer.step()

            # Calculate metrics
            with torch.no_grad():
                fused_np = fused_output.cpu().numpy()[0, 0]
                gt_np = ground_truth.cpu().numpy()[0, 0]

                batch_ssim = ssim(fused_np, gt_np, data_range=1.0)
                batch_psnr = psnr(gt_np, fused_np, data_range=1.0)

                epoch_loss += loss.item()
                epoch_ssim += batch_ssim
                epoch_psnr += batch_psnr
                num_batches += 1

        # Update learning rate
        scheduler.step()

        # Calculate averages
        avg_loss = epoch_loss / num_batches
        avg_ssim = epoch_ssim / num_batches
        avg_psnr = epoch_psnr / num_batches

        training_history['loss'].append(avg_loss)
        training_history['ssim'].append(avg_ssim)
        training_history['psnr'].append(avg_psnr)

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.6f}, SSIM: {avg_ssim:.4f}, PSNR: {avg_psnr:.2f} dB')

    return training_history

def extract_radiomic_features(image):
    """Extract radiomic features for enhanced classification"""
    features = {}

    # First-order statistics
    features['mean'] = np.mean(image)
    features['std'] = np.std(image)
    features['skewness'] = float(ndi.measurements.center_of_mass(image)[0]) if image.sum() > 0 else 0
    features['entropy'] = -np.sum(image * np.log2(image + 1e-10))

    # Texture features (simplified GLCM-like)
    features['contrast'] = np.var(image)
    features['homogeneity'] = 1.0 / (1.0 + np.var(image))

    # Shape features
    binary_img = (image > np.mean(image)).astype(int)
    features['area'] = np.sum(binary_img)
    features['compactness'] = features['area'] / (np.sum(binary_img > 0) + 1e-10)

    return features

def train_classification_model(fusion_model, dataloader, num_epochs=20):
    """Train classification model on fused images"""
    classifier = MetastasesClassifier(num_classes=2).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(classifier.parameters(), lr=0.0001)

    fusion_model.eval()  # Set fusion model to evaluation mode
    classifier.train()

    training_history = {'loss': [], 'accuracy': []}

    for epoch in range(num_epochs):
        epoch_loss = 0.0
        correct_predictions = 0
        total_predictions = 0

        for batch_idx, (ct_batch, mri_batch, labels) in enumerate(dataloader):
            ct_batch, mri_batch = ct_batch.to(device), mri_batch.to(device)
            labels = labels.to(device).squeeze()

            # Ensure labels have the correct shape
            if labels.dim() == 0:
                labels = labels.unsqueeze(0)

            # Generate fused image
            with torch.no_grad():
                fused_img = fusion_model(ct_batch, mri_batch)

            # Convert to 3-channel for VGG19
            fused_img_3ch = fused_img.repeat(1, 3, 1, 1)

            # Forward pass through classifier
            optimizer.zero_grad()
            outputs = classifier(fused_img_3ch)
            loss = criterion(outputs, labels)

            # Backward pass
            loss.backward()
            optimizer.step()

            # Calculate accuracy
            _, predicted = torch.max(outputs.data, 1)
            total_predictions += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()
            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(dataloader)
        accuracy = 100 * correct_predictions / total_predictions

        training_history['loss'].append(avg_loss)
        training_history['accuracy'].append(accuracy)

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}, Accuracy: {accuracy:.2f}%')

    return classifier, training_history

# ==================== EVALUATION FUNCTIONS ====================

def evaluate_fusion_quality(fused_img, ct_img, mri_img):
    """Evaluate fusion quality using multiple metrics"""
    metrics = {}

    # Convert to numpy if tensor
    if torch.is_tensor(fused_img):
        fused_img = fused_img.cpu().numpy()
    if torch.is_tensor(ct_img):
        ct_img = ct_img.cpu().numpy()
    if torch.is_tensor(mri_img):
        mri_img = mri_img.cpu().numpy()

    # Reference image (average)
    reference = (ct_img + mri_img) / 2

    # SSIM
    metrics['ssim'] = ssim(fused_img, reference, data_range=1.0)

    # PSNR
    metrics['psnr'] = psnr(reference, fused_img, data_range=1.0)

    # Entropy
    fused_img_normalized = fused_img / (fused_img.sum() + 1e-10)
    metrics['entropy'] = -np.sum(fused_img_normalized * np.log2(fused_img_normalized + 1e-10))

    # Edge preservation index
    ct_edges = cv2.Canny((ct_img * 255).astype(np.uint8), 50, 150)
    mri_edges = cv2.Canny((mri_img * 255).astype(np.uint8), 50, 150)
    fused_edges = cv2.Canny((fused_img * 255).astype(np.uint8), 50, 150)

    total_edges = np.logical_or(ct_edges, mri_edges).sum()
    preserved_edges = np.logical_and(fused_edges, np.logical_or(ct_edges, mri_edges)).sum()
    metrics['edge_preservation'] = preserved_edges / (total_edges + 1e-10)

    return metrics

def detect_metastases_regions(fused_img, threshold=0.8):
    """Detect potential metastases regions in fused image"""
    # Apply threshold to identify bright regions
    binary = (fused_img > threshold).astype(np.uint8)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    detections = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 10 and area < 500:  # Filter by size
            x, y, w, h = cv2.boundingRect(contour)
            confidence = np.mean(fused_img[y:y+h, x:x+w])

            detections.append({
                'bbox': (x, y, w, h),
                'area': area,
                'confidence': confidence
            })

    return detections


# ==================== VISUALIZATION FUNCTIONS ====================

def visualize_fusion_results(ct_img, mri_img, fused_img, detections=None):
    """Visualize fusion results with potential metastases detection"""
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Original images
    axes[0, 0].imshow(ct_img, cmap='gray')
    axes[0, 0].set_title('CT Image')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(mri_img, cmap='gray')
    axes[0, 1].set_title('MRI Image')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(fused_img, cmap='gray')
    axes[0, 2].set_title('Fused Image')
    axes[0, 2].axis('off')

    # Enhanced visualizations
    axes[1, 0].imshow(ct_img, cmap='bone')
    axes[1, 0].set_title('CT (Bone Window)')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(mri_img, cmap='viridis')
    axes[1, 1].set_title('MRI (Enhanced)')
    axes[1, 1].axis('off')

    # Fused with detections
    axes[1, 2].imshow(fused_img, cmap='gray')
    if detections:
        for detection in detections:
            x, y, w, h = detection['bbox']
            rect = plt.Rectangle((x, y), w, h, linewidth=2,
                               edgecolor='red', facecolor='none')
            axes[1, 2].add_patch(rect)
            axes[1, 2].text(x, y-2, f'Conf: {detection["confidence"]:.2f}',
                           color='red', fontsize=8)
    axes[1, 2].set_title('Fued with Detections')
    axes[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

def plot_training_history(fusion_history, classification_history=None):
    """Plot training histories"""
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # Fusion model history
    axes[0, 0].plot(fusion_history['loss'])
    axes[0, 0].set_title('Fusion Model - Training Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('MSE Loss')

    axes[0, 1].plot(fusion_history['ssim'])
    axes[0, 1].set_title('Fusion Model - SSIM')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('SSIM')

    axes[0, 2].plot(fusion_history['psnr'])
    axes[0, 2].set_title('Fusion Model - PSNR')
    axes[0, 2].set_xlabel('Epoch')
    axes[0, 2].set_ylabel('PSNR (dB)')

    if classification_history:
        axes[1, 0].plot(classification_history['loss'])
        axes[1, 0].set_title('Classification Model - Training Loss')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('CrossEntropy Loss')

        axes[1, 1].plot(classification_history['accuracy'])
        axes[1, 1].set_title('Classification Model - Accuracy')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].set_ylabel('Accuracy (%)')

    plt.tight_layout()
    plt.show()

# ==================== ENHANCED PREPROCESSING AND VISUALIZATION ====================

def enhanced_preprocess_medical_image(image, target_size=(256, 256), clip_limit=3.0):
    """
    Enhanced preprocessing for medical images with better contrast normalization
    and noise reduction
    """
    # Normalize to 0-1 range
    image = image.astype(np.float32)
    if np.max(image) > np.min(image):
        image = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Apply CLAHE for contrast enhancement (on 8-bit representation)
    image_8bit = (image * 255).astype(np.uint8)
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=(8, 8))
    image_enhanced = clahe.apply(image_8bit)

    # Apply adaptive histogram equalization
    image_enhanced = cv2.equalizeHist(image_enhanced)

    # Apply non-local means denoising
    image_denoised = cv2.fastNlMeansDenoising(image_enhanced, h=10)

    # Resize to target size with high-quality interpolation
    image_resized = cv2.resize(image_denoised, target_size, interpolation=cv2.INTER_CUBIC)

    # Convert back to float and normalize
    image_processed = image_resized.astype(np.float32) / 255.0

    return image_processed

def apply_medical_windowing(image, window_center=0.5, window_width=1.0):
    """
    Apply medical-style windowing to enhance visualization
    """
    # Calculate window boundaries
    low = max(0, window_center - window_width/2)
    high = min(1, window_center + window_width/2)

    # Apply windowing
    windowed = np.clip((image - low) / (high - low), 0, 1)

    return windowed

# ==================== 3D VISUALIZATION FUNCTIONS ====================

def visualize_3d_brain_volume(volume, title="3D Brain Volume", threshold=0.5):
    """
    Create a 3D visualization of a brain volume using matplotlib
    """
    try:
        from mpl_toolkits.mplot3d import Axes3D
        from skimage import measure

        # Create a figure
        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')

        # Extract surfaces using marching cubes
        verts, faces, normals, values = measure.marching_cubes(volume, threshold)

        # Plot the surface
        ax.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                       cmap='Spectral', alpha=0.7, linewidth=0.1)

        ax.set_title(title)
        ax.set_xlabel('X-axis')
        ax.set_ylabel('Y-axis')
        ax.set_ylabel('Z-axis')

        plt.tight_layout()
        return fig
    except ImportError:
        print("3D visualization requires scikit-image and 3D matplotlib support")
        return None
    except Exception as e:
        print(f"3D visualization error: {e}")
        return None

def create_3d_metastases_visualization(ct_volume, mri_volume, fused_volume, detections_3d):
    """
    Create a comprehensive 3D visualization of brain with detected metastases
    """
    fig = plt.figure(figsize=(15, 10))

    # Plot CT volume
    ax1 = fig.add_subplot(221, projection='3d')
    try:
        from skimage import measure
        verts, faces, _, _ = measure.marching_cubes(ct_volume, 0.5)
        ax1.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                        cmap='gray', alpha=0.7)
        ax1.set_title('CT Volume')
    except:
        ax1.text(0.5, 0.5, 0.5, "3D CT Visualization\nNot Available",
                ha='center', va='center', transform=ax1.transAxes)

    # Plot MRI volume
    ax2 = fig.add_subplot(222, projection='3d')
    try:
        verts, faces, _, _ = measure.marching_cubes(mri_volume, 0.5)
        ax2.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                        cmap='viridis', alpha=0.7)
        ax2.set_title('MRI Volume')
    except:
        ax2.text(0.5, 0.5, 0.5, "3D MRI Visualization\nNot Available",
                ha='center', va='center', transform=ax2.transAxes)

    # Plot fused volume
    ax3 = fig.add_subplot(223, projection='3d')
    try:
        verts, faces, _, _ = measure.marching_cubes(fused_volume, 0.5)
        ax3.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                        cmap='plasma', alpha=0.7)
        ax3.set_title('Fused Volume')
    except:
        ax3.text(0.5, 0.5, 0.5, "3D Fused Visualization\nNot Available",
                ha='center', va='center', transform=ax3.transAxes)

    # Plot with detections
    ax4 = fig.add_subplot(224, projection='3d')
    try:
        # Plot the brain surface
        verts, faces, _, _ = measure.marching_cubes(fused_volume, 0.5)
        ax4.plot_trisurf(verts[:, 0], verts[:, 1], faces, verts[:, 2],
                        color='lightgray', alpha=0.3)

        # Plot detected metastases
        for detection in detections_3d:
            x, y, z = detection['center']
            size = detection['size']

            # Create a sphere for the detection
            u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
            x_sphere = x + size * np.cos(u) * np.sin(v)
            y_sphere = y + size * np.sin(u) * np.sin(v)
            z_sphere = z + size * np.cos(v)

            ax4.plot_surface(x_sphere, y_sphere, z_sphere, color='red', alpha=0.7)

        ax4.set_title('Detected Metastases (3D)')
    except:
        ax4.text(0.5, 0.5, 0.5, "3D Detection Visualization\nNot Available",
                ha='center', va='center', transform=ax4.transAxes)

    plt.tight_layout()
    return fig

# ==================== ENHANCED DETECTION ALGORITHMS ====================

def enhanced_detect_metastases_regions(fused_img, ct_img=None, mri_img=None,
                                     threshold=0.7, min_size=10, max_size=500):
    """
    Enhanced metastases detection using multi-modal information and
    morphological processing to reduce false positives
    """
    # Create a multi-modal detection mask
    if ct_img is not None and mri_img is not None:
        # Combine information from all modalities
        ct_mask = (ct_img > np.percentile(ct_img, 75)).astype(np.uint8)
        mri_mask = (mri_img > np.percentile(mri_img, 75)).astype(np.uint8)
        fused_mask = (fused_img > threshold).astype(np.uint8)

        # Combine masks with logical AND to reduce false positives
        combined_mask = np.logical_and(fused_mask, np.logical_or(ct_mask, mri_mask)).astype(np.uint8)
    else:
        # Fallback to single modality
        combined_mask = (fused_img > threshold).astype(np.uint8)

    # Apply morphological operations to clean up the mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_CLOSE, kernel)
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_OPEN, kernel)

    # Find contours
    contours, _ = cv2.findContours(combined_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    detections = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if min_size <= area <= max_size:  # Filter by size
            x, y, w, h = cv2.boundingRect(contour)

            # Calculate confidence based on intensity and shape
            region_intensity = np.mean(fused_img[y:y+h, x:x+w])
            circularity = 4 * np.pi * area / (cv2.arcLength(contour, True)**2) if area > 0 else 0

            # Higher confidence for circular regions with high intensity
            confidence = region_intensity * (0.5 + 0.5 * circularity)

            detections.append({
                'bbox': (x, y, w, h),
                'area': area,
                'confidence': confidence,
                'circularity': circularity
            })

    # Sort by confidence (highest first)
    detections.sort(key=lambda x: x['confidence'], reverse=True)

    return detections

def detect_3d_metastases(volume, threshold=0.7, min_volume=50, max_volume=1000):
    """
    Detect metastases in 3D volumes
    """
    try:
        from skimage import measure

        # Create binary mask
        binary_volume = volume > threshold

        # Label connected components
        labeled_volume = measure.label(binary_volume)

        # Get region properties
        regions = measure.regionprops(labeled_volume, intensity_image=volume)

        detections = []
        for region in regions:
            if min_volume <= region.area <= max_volume:
                # Calculate confidence based on intensity and sphericity
                mean_intensity = np.mean(region.intensity_image)
                sphericity = calculate_sphericity(region)

                confidence = mean_intensity * (0.5 + 0.5 * sphericity)

                detections.append({
                    'center': region.centroid,
                    'size': np.mean(region.bbox[3:] - region.bbox[:3]),
                    'volume': region.area,
                    'confidence': confidence,
                    'sphericity': sphericity
                })

        # Sort by confidence
        detections.sort(key=lambda x: x['confidence'], reverse=True)
        return detections

    except ImportError:
        print("3D detection requires scikit-image")
        return []

def calculate_sphericity(region):
    """Calculate sphericity of a 3D region"""
    try:
        # Surface area of a sphere with the same volume
        volume = region.area
        sphere_surface = (36 * np.pi * volume**2) ** (1/3)

        # Actual surface area (approximate)
        surface_area = region.area  # This is a simplification

        # Sphericity ratio (1 for perfect sphere)
        sphericity = sphere_surface / surface_area if surface_area > 0 else 0

        return min(1.0, sphericity)
    except:
        return 0.5  # Default value

# ==================== ENHANCED VISUALIZATION FUNCTIONS ====================

def enhanced_visualize_medical_fusion_results(ct_img, mri_img, fused_img, detections=None,
                                            title_suffix="", save_path=None):
    """
    Enhanced visualization for medical data with better contrast and annotations
    """
    # Apply medical windowing for better visualization
    ct_windowed = apply_medical_windowing(ct_img, window_center=0.4, window_width=0.8)
    mri_windowed = apply_medical_windowing(mri_img, window_center=0.5, window_width=0.8)
    fused_windowed = apply_medical_windowing(fused_img, window_center=0.5, window_width=0.8)

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    # Original modalities with enhanced contrast
    axes[0, 0].imshow(ct_windowed, cmap='gray')
    axes[0, 0].set_title(f'T1Gd (Contrast Enhanced) {title_suffix}', fontweight='bold')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(mri_windowed, cmap='gray')
    axes[0, 1].set_title(f'T2 Weighted {title_suffix}', fontweight='bold')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(fused_windowed, cmap='gray')
    axes[0, 2].set_title(f'AI Fused Image {title_suffix}', fontweight='bold')
    axes[0, 2].axis('off')

    # Enhanced views with color mapping
    axes[1, 0].imshow(ct_windowed, cmap='hot')
    axes[1, 0].set_title('T1Gd (Hot Colormap)', fontweight='bold')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(mri_windowed, cmap='viridis')
    axes[1, 1].set_title('T2 (Enhanced View)', fontweight='bold')
    axes[1, 1].axis('off')

    # Fused with AI detections
    axes[1, 2].imshow(fused_windowed, cmap='gray')
    if detections:
        for i, detection in enumerate(detections):
            x, y, w, h = detection['bbox']
            rect = plt.Rectangle((x, y), w, h, linewidth=2,
                               edgecolor='red', facecolor='none')
            axes[1, 2].add_patch(rect)
            axes[1, 2].text(x, y-5, f'ROI-{i+1}\n({detection["confidence"]:.2f})',
                           color='red', fontsize=8, fontweight='bold',
                           bbox=dict(facecolor='white', alpha=0.7, edgecolor='red'))
    axes[1, 2].set_title('AI Analysis + ROI Detection', fontweight='bold')
    axes[1, 2].axis('off')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Visualization saved to: {save_path}")

    plt.show()

# ==================== INTEGRATION WITH EXISTING CODE ====================

# Update the MedicalDecathlonBrainDataset class with enhanced preprocessing
class EnhancedMedicalDecathlonDataset(MedicalDecathlonBrainDataset):
    """Enhanced dataset with better preprocessing"""

    def preprocess_image(self, img):
        """Enhanced preprocessing for medical images"""
        return enhanced_preprocess_medical_image(img, self.target_size)

    def load_nifti_volume_3d(self, filepath):
        """Load full 3D volume for 3D processing"""
        try:
            nii_img = nib.load(filepath)
            volume = nii_img.get_fdata()

            # Normalize the volume
            volume = volume.astype(np.float32)
            volume_min, volume_max = volume.min(), volume.max()
            if volume_max > volume_min:
                volume = (volume - volume_min) / (volume_max - volume_min)

            return volume
        except Exception as e:
            print(f"Error loading 3D volume: {e}")
            return None

# Update the main function to use enhanced methods
def enhanced_main_with_real_data():
    """Enhanced main function with better visualization and 3D support"""
    print("=== Enhanced Brain Metastases Detection with 3D Support ===")

    # Setup dataset with enhanced preprocessing
    data_root = './medical_decathlon_data/Task01_BrainTumour'

    if not os.path.exists(data_root):
        print("Medical Decathlon dataset not found.")
        return

    # Create enhanced dataset
    enhanced_dataset = EnhancedMedicalDecathlonDataset(data_root, split='training')

    # Load models
    fusion_model = EnhancedFusionAutoencoder().to(device)
    classifier = MetastasesClassifier().to(device)

    # Load pre-trained weights if available
    fusion_model_path = 'brain_fusion_model.pth'
    classifier_path = 'metastases_classifier.pth'

    if os.path.exists(fusion_model_path) and os.path.exists(classifier_path):
        fusion_model.load_state_dict(torch.load(fusion_model_path, map_location=device))
        classifier.load_state_dict(torch.load(classifier_path, map_location=device))
        print("Loaded pre-trained models")
    else:
        print("Pre-trained models not found")
        return

    # Process a case with enhanced visualization
    case_index = 0  # You can change this to view different cases
    ct_tensor, mri_tensor, label = enhanced_dataset[case_index]

    # Generate fused image
    fusion_model.eval()
    with torch.no_grad():
        fused_img = fusion_model(ct_tensor.unsqueeze(0).to(device),
                               mri_tensor.unsqueeze(0).to(device))
        fused_img = fused_img[0, 0].cpu().numpy()

    # Enhanced detection
    ct_np = ct_tensor[0].cpu().numpy()
    mri_np = mri_tensor[0].cpu().numpy()
    detections = enhanced_detect_metastases_regions(fused_img, ct_np, mri_np)

    # Enhanced visualization
    enhanced_visualize_medical_fusion_results(
        ct_np, mri_np, fused_img, detections,
        title_suffix=f"Real Medical Sample {case_index+1}"
    )

    # Try 3D visualization if data is available
    try:
        # Load 3D volume
        volume_path = os.path.join(data_root, 'imagesTr', enhanced_dataset.image_files[case_index])
        volume_3d = enhanced_dataset.load_nifti_volume_3d(volume_path)

        if volume_3d is not None:
            # For demonstration, we'll use a simple approach to 3D visualization
            # In a real application, you would process the entire volume

            # Extract a sub-volume for visualization
            z_start, z_end = volume_3d.shape[2]//2 - 10, volume_3d.shape[2]//2 + 10
            sub_volume = volume_3d[:, :, z_start:z_end]

            # Create 3D visualization
            fig_3d = visualize_3d_brain_volume(sub_volume, "3D Brain Volume (Partial)")
            if fig_3d:
                plt.show()
    except Exception as e:
        print(f"3D visualization not available: {e}")


# Add these enhanced functions to your existing code

# ==================== ENHANCED PREPROCESSING AND VISUALIZATION ====================

def enhanced_preprocess_medical_image(image, target_size=(256, 256), clip_limit=3.0):
    """
    Enhanced preprocessing for medical images with better contrast normalization
    and noise reduction
    """
    # Normalize to 0-1 range
    image = image.astype(np.float32)
    if np.max(image) > np.min(image):
        image = (image - np.min(image)) / (np.max(image) - np.min(image))

    # Apply CLAHE for contrast enhancement (on 8-bit representation)
    image_8bit = (image * 255).astype(np.uint8)
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=(8, 8))
    image_enhanced = clahe.apply(image_8bit)

    # Apply adaptive histogram equalization
    image_enhanced = cv2.equalizeHist(image_enhanced)

    # Apply non-local means denoising
    image_denoised = cv2.fastNlMeansDenoising(image_enhanced, h=10)

    # Resize to target size with high-quality interpolation
    image_resized = cv2.resize(image_denoised, target_size, interpolation=cv2.INTER_CUBIC)

    # Convert back to float and normalize
    image_processed = image_resized.astype(np.float32) / 255.0

    return image_processed

def apply_medical_windowing(image, window_center=0.5, window_width=1.0):
    """
    Apply medical-style windowing to enhance visualization
    """
    # Calculate window boundaries
    low = max(0, window_center - window_width/2)
    high = min(1, window_center + window_width/2)

    # Apply windowing
    windowed = np.clip((image - low) / (high - low), 0, 1)

    return windowed

def enhanced_detect_metastases_regions(fused_img, ct_img=None, mri_img=None,
                                     threshold=0.7, min_size=10, max_size=500):
    """
    Enhanced metastases detection using multi-modal information and
    morphological processing to reduce false positives
    """
    # Create a multi-modal detection mask
    if ct_img is not None and mri_img is not None:
        # Combine information from all modalities
        ct_mask = (ct_img > np.percentile(ct_img, 75)).astype(np.uint8)
        mri_mask = (mri_img > np.percentile(mri_img, 75)).astype(np.uint8)
        fused_mask = (fused_img > threshold).astype(np.uint8)

        # Combine masks with logical AND to reduce false positives
        combined_mask = np.logical_and(fused_mask, np.logical_or(ct_mask, mri_mask)).astype(np.uint8)
    else:
        # Fallback to single modality
        combined_mask = (fused_img > threshold).astype(np.uint8)

    # Apply morphological operations to clean up the mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_CLOSE, kernel)
    combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_OPEN, kernel)

    # Find contours
    contours, _ = cv2.findContours(combined_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    detections = []
    for contour in contours:
        area = cv2.contourArea(contour)
        if min_size <= area <= max_size:  # Filter by size
            x, y, w, h = cv2.boundingRect(contour)

            # Calculate confidence based on intensity and shape
            region_intensity = np.mean(fused_img[y:y+h, x:x+w])
            circularity = 4 * np.pi * area / (cv2.arcLength(contour, True)**2) if area > 0 else 0

            # Higher confidence for circular regions with high intensity
            confidence = region_intensity * (0.5 + 0.5 * circularity)

            detections.append({
                'bbox': (x, y, w, h),
                'area': area,
                'confidence': confidence,
                'circularity': circularity
            })

    # Sort by confidence (highest first)
    detections.sort(key=lambda x: x['confidence'], reverse=True)

    return detections

def enhanced_visualize_medical_fusion_results(ct_img, mri_img, fused_img, detections=None,
                                            title_suffix="", save_path=None):
    """
    Enhanced visualization for medical data with better contrast and annotations
    """
    # Apply medical windowing for better visualization
    ct_windowed = apply_medical_windowing(ct_img, window_center=0.4, window_width=0.8)
    mri_windowed = apply_medical_windowing(mri_img, window_center=0.5, window_width=0.8)
    fused_windowed = apply_medical_windowing(fused_img, window_center=0.5, window_width=0.8)

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    # Original modalities with enhanced contrast
    axes[0, 0].imshow(ct_windowed, cmap='gray')
    axes[0, 0].set_title(f'T1Gd (Contrast Enhanced) {title_suffix}', fontweight='bold')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(mri_windowed, cmap='gray')
    axes[0, 1].set_title(f'T2 Weighted {title_suffix}', fontweight='bold')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(fused_windowed, cmap='gray')
    axes[0, 2].set_title(f'AI Fused Image {title_suffix}', fontweight='bold')
    axes[0, 2].axis('off')

    # Enhanced views with color mapping
    axes[1, 0].imshow(ct_windowed, cmap='hot')
    axes[1, 0].set_title('T1Gd (Hot Colormap)', fontweight='bold')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(mri_windowed, cmap='viridis')
    axes[1, 1].set_title('T2 (Enhanced View)', fontweight='bold')
    axes[1, 1].axis('off')

    # Fused with AI detections
    axes[1, 2].imshow(fused_windowed, cmap='gray')
    if detections:
        for i, detection in enumerate(detections):
            x, y, w, h = detection['bbox']
            rect = plt.Rectangle((x, y), w, h, linewidth=2,
                               edgecolor='red', facecolor='none')
            axes[1, 2].add_patch(rect)
            axes[1, 2].text(x, y-5, f'ROI-{i+1}\n({detection["confidence"]:.2f})',
                           color='red', fontsize=8, fontweight='bold',
                           bbox=dict(facecolor='white', alpha=0.7, edgecolor='red'))
    axes[1, 2].set_title('AI Analysis + ROI Detection', fontweight='bold')
    axes[1, 2].axis('off')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Visualization saved to: {save_path}")

    plt.show()

# ==================== ENHANCED MAIN FUNCTION ====================

def enhanced_main_with_real_data():
    """Enhanced main function with better visualization"""
    print("=== Enhanced Brain Metastases Detection with Improved Visualization ===")

    # Use the original dataset setup function
    data_root = './medical_decathlon_data/Task01_BrainTumour'

    if not os.path.exists(data_root):
        print("Medical Decathlon dataset not found. Starting download process...")
        data_root = download_medical_decathlon_brain_data()

        if data_root is None:
            print("Please download the dataset manually and rerun.")
            return

    # Setup datasets using the original function
    train_dataset, val_dataset, fusion_dataset = setup_medical_decathlon_dataset(data_root)

    if train_dataset is None:
        print("Failed to setup dataset. Please check the data structure.")
        return

    # Load models
    fusion_model = EnhancedFusionAutoencoder().to(device)
    classifier = MetastasesClassifier().to(device)

    # Load pre-trained weights if available
    fusion_model_path = 'brain_fusion_model.pth'
    classifier_path = 'metastases_classifier.pth'

    if os.path.exists(fusion_model_path) and os.path.exists(classifier_path):
        fusion_model.load_state_dict(torch.load(fusion_model_path, map_location=device))
        classifier.load_state_dict(torch.load(classifier_path, map_location=device))
        print("Loaded pre-trained models")
    else:
        print("Pre-trained models not found. Please train models first.")
        return

    # Process a case with enhanced visualization
    case_index = 0  # You can change this to view different cases

    # Get the case data
    ct_tensor, mri_tensor, label = val_dataset[case_index]

    # Generate fused image
    fusion_model.eval()
    with torch.no_grad():
        fused_img = fusion_model(ct_tensor.unsqueeze(0).to(device),
                               mri_tensor.unsqueeze(0).to(device))
        fused_img = fused_img[0, 0].cpu().numpy()

    # Enhanced detection
    ct_np = ct_tensor[0].cpu().numpy()
    mri_np = mri_tensor[0].cpu().numpy()
    detections = enhanced_detect_metastases_regions(fused_img, ct_np, mri_np)

    # Enhanced visualization
    enhanced_visualize_medical_fusion_results(
        ct_np, mri_np, fused_img, detections,
        title_suffix=f"Real Medical Sample {case_index+1}"
    )

    # Print case information
    print(f"\nCase Information:")
    print(f"Ground Truth: {'Tumor Present' if label.item() == 1 else 'No Tumor'}")
    print(f"Number of Detected Regions: {len(detections)}")

    if detections:
        print("\nDetected Regions:")
        for i, detection in enumerate(detections):
            print(f"Region {i+1}: Confidence={detection['confidence']:.3f}, Area={detection['area']}")

# ==================== EXECUTION ====================

if __name__ == "__main__":
    # Set random seeds for reproducibility
    torch.manual_seed(42)
    np.random.seed(42)

    print("ENHANCED BRAIN METASTASES DETECTION PROJECT")
    print("==========================================")
    print("Choose execution mode:")
    print("1. Original training and evaluation")
    print("2. Enhanced visualization with real data")

    choice = input("Enter choice (1-2): ").strip()

    if choice == '1':
        print("\nRunning original training and evaluation...")
        main_with_real_data()
    elif choice == '2':
        print("\nRunning enhanced visualization with real data...")
        enhanced_main_with_real_data()
    else:
        print("Invalid choice!")

```
